```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dummy sales data
data = {
    "Month": pd.date_range(start="2022-01-01", periods=24, freq="ME"),
    "Sales": [1200, 1350, 1500, 1450, 1600, 1700, 1650, 1800,
              1900, 2000, 2100, 2200, 2300, 2400, 2500, 2600,
              2550, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Month"], df["Sales"], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
df["Forecast"] = df["Sales"].rolling(window=3).mean()
df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Month"], df["Sales"], label="Actual Sales")
plt.plot(df["Month"], df["Forecast"], label="Forecast", linestyle="--")
plt.legend()
plt.title("Sales Forecast")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
